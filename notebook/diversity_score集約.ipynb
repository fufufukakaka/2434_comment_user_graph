```python
import pandas as pd
```

```python
diversity_df = pd.read_csv("../diversity_df.csv")
concat_video_list = pd.read_csv("../concat_video_list.csv")
```

```python
diversity_df.tail()
```

```python
streamers = diversity_df.columns[1:-1]
```

```python
streamers
```

```python
concat_video_list.head()
```

```python
merged = pd.merge(concat_video_list, diversity_df)
```

```python
merged.tail(10)
```

```python
merged["diversity_score"] = merged.apply(lambda x: x[[v for v in streamers if v != x["streamer_name"]]].sum(), axis=1)
```

```python
merged.head()
```

```python
merged.sort_values("diversity_score", ascending=False)
```

```python
merged.sort_values("diversity_score", ascending=False).to_csv("../diversity_score.csv", index=None)
```

```python
# 配信者ごとにどの配信者から流入しているかをgroupby sumする
grouped_sum = merged[["streamer_name"]+list(streamers)+["sum_score"]].groupby("streamer_name").sum()
```

```python
grouped_sum.head()
```

```python
# 自分のところでコメントしているユーザは、どれくらい他のところでもコメントしているのか、の割合
grouped_sum = grouped_sum.reset_index()
grouped_sum["also_chat_others_ratio"] = grouped_sum.apply(lambda x: (x["sum_score"]-x[x["streamer_name"]])/x["sum_score"], axis=1)
```

```python
grouped_sum.sort_values("also_chat_others_ratio", ascending=True).head(10)
```

```python
grouped_sum[["streamer_name", "also_chat_others_ratio"]].to_csv("../also_chat_others_ratio.csv", index=None)
```
